### Questão 13
Calcule a **média de `valor_parcela` por UF** e informe a maior média encontrada. Responda com **um número decimal**.


```python
df['VALOR PARCELA_num'] = df['VALOR PARCELA'].str.replace(',', '.', regex=False).astype(float)
df_uf_media = df.groupby('UF')['VALOR PARCELA_num'].mean()
print('%.2f'%df_uf_media.max())
```

### Questão 14
Calcule a **mediana de `valor_parcela` por município** e informe a maior mediana encontrada. Responda com **um número decimal**.


```python
df_mun_mediana = df.groupby('NOME MUNICÍPIO')['VALOR PARCELA_num'].median()
print('%.2f'%df_mun_mediana.max())
```

### Questão 15
Para cada `mes_referencia`, calcule o **máximo** de `valor_parcela`. Informe o maior máximo encontrado. Responda com **um número decimal**.


```python
df_mes_max = df.groupby('MÊS REFERÊNCIA')['VALOR PARCELA_num'].max()
print('%.2f'%df_mes_max.max())
```
